```python
def get_stations_by_indicator(indicator):
    ind_posts = wroc_working_posts[wroc_working_posts['Wskaźnik - kod'] == indicator]['Kod stacji']
    ind_posts_meta = metadata[metadata['Kod stacji'].isin(ind_posts)]
    m = folium.Map(location=[51.1079,17.0385], zoom_start=12)
    for index, row in ind_posts_meta.iterrows():
        folium.Marker([row['WGS84 φ N'], row['WGS84 λ E']], popup=row['Nazwa stacji']).add_to(m)
    return m, ind_posts
```

```python
# 

column_mapper = {row[1]: row[0] for _, row in metadata[['Kod stacji', 'Stary Kod stacji']].iterrows()}


# excel format changes across years, some of the couple of first rows should be ignored
skips = {
    2014: [1, 2], 
    2015: [1, 2],
    2016: [0, 2, 3, 4, 5], 
    2017: [0, 2, 3, 4, 5],
    2018: [0, 2, 3, 4, 5],
    2019: [0, 2, 3, 4, 5],
    2020: [0, 2, 3, 4, 5],
}

def read_skipped(year, indicator, sampling_time):
    path = os.path.join('..', 'data', str(year), f'{year}_{indicator}_{sampling_time}.xlsx')
    data = pd.read_excel(path, skiprows=skips[year])
    data = data.rename(columns={ data.columns[0]: "timestamp" })
    data = data.rename(columns=column_mapper)
    data = data.set_index('timestamp')
    # some numbers are entered with a comma instead of a dot
    data = data.applymap(lambda x: str(x).replace(',', '.')).astype(float)
    return data
```

```python
def collect_yearly_data(indicator):
    stations = get_stations_by_indicator(indicator)[1]
    all_years = []
    # we deliberately skip 2014 as it is full of missing values and even entire stations although metadata shows data should be there 
    for year in range(2015, 2021):
        # by default, try to load 24g data
        y_stations = set(stations)
        complete_year_data_24 = read_skipped(year, indicator, '24g')
        # see which stations do not have 24g sampling
        y_stat_24g = y_stations.intersection(set(complete_year_data_24.columns))
        y_stat_1g = y_stations.difference(set(complete_year_data_24.columns))

        data = complete_year_data_24[y_stat_24g]
        # get rest of station with 1g sampling and resample to 24g
        if y_stat_1g:
            complete_year_data_1 = read_skipped(year, indicator, '1g').resample('1D').mean()
            complete_year_data_1 = complete_year_data_1.drop(
                pd.date_range(f'{year+1}-01-01', f'{year+1}-02-01'), errors='ignore')
            data_1g = complete_year_data_1[y_stat_1g]

            data = pd.concat([data, data_1g], axis=1)
        all_years.append(data)
    
    re = pd.concat(all_years)
    print("Before interpolating: ", re.isnull().sum(axis = 0))
    re = re.interpolate()
    return re
    
        
```

```python
no2_data = collect_yearly_data('NO2')
```

```python
pm25_data = collect_yearly_data('PM25')
```
